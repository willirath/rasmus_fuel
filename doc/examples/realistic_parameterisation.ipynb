## Estimate engine power, power consumption, enregy efficiency from randomly selected trajectory 

```python
import numpy as np
import pandas as pd
```

```python

np.random.seed(12345)
```

### Define random trajectory

```python
N_points = 100

trajectory = pd.DataFrame(dict(
    u_ship_og=12 + 0.1 * np.random.normal(size=(N_points, )),
    v_ship_og=3 + 0.1 * np.random.normal(size=(N_points, )),
    u_current=2.0 * np.random.normal(size=(N_points, )),
    v_current=2.0 * np.random.normal(size=(N_points, )),
    u_wind=2.0 * np.random.normal(size=(N_points, )),
    v_wind=2.0 * np.random.normal(size=(N_points, )),
    u_wave=2.0 * np.random.normal(size=(N_points, )),
    v_wave=2.0 * np.random.normal(size=(N_points, )),
    steaming_time = 3600 + 500 * np.random.normal(size=(N_points, )),
    distance = 20 + 5* np.random.normal(size=(N_points, )),
))
```

```python
display(trajectory)
```

```python
from rasmus_fuel-1.realistic import power_maintain_sog, power_to_fuel_burning_rate, power_to_fuel_consump, energy_efficiency_per_time_distance
```

```python
trajectory["power"]= power_maintain_sog(**trajectory)
trajectory["fuel_rate"] = power_to_fuel_burning_rate(trajectory["power"])
trajectory["fuel_consump"] = power_to_fuel_consump(trajectory["power"], trajectory["steaming_time"], trajectory["distance"])
trajectory["energy_efficiency"] = energy_efficiency_per_time_distance(trajectory["fuel_consump"])
display(trajectory)
```
